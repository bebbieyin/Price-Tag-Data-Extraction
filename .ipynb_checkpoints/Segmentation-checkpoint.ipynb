```python
import torch
from PIL import Image
from torchvision import transforms
import cv2
import matplotlib.pyplot as plt
import numpy as np
from torch import nn
import os
```

```python
def segment(img):
    img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
    img = cv2.resize(img,(224,224))
    img1 = img.transpose(2,0,1)
    img1 = img1.reshape(1,3,224,224)
    with torch.no_grad():
        a = model(torch.from_numpy(img1)/255)
    output = a.cpu().detach().numpy()[0][0]>0.3
    return output
    
# expand the white area of the original horizonal area of getting the min y value and max y value
def get_corners(mask_img):
    coords = []
    for ycoord,i in enumerate(mask_img):
        for xcoord,j in enumerate(i):
            if j==True:
                coords.append([xcoord,ycoord])

    coords_arr = np.array(coords)
    xmax, ymax = coords_arr.max(axis=0)
    xmin,ymin = coords_arr.min(axis=0)
    return ymax,ymin

# create the new mask based on the min y and max y, optionally can choose to add in some space
def create_wider_mask(old_mask,ymax,ymin,extend_val=0):
    
    old_mask[ymin-extend_val:ymax+extend_val] = True
    

# crop out the mykad based on the mask's white values
def get_segment_crop(img,tol=0, mask=None):
    if mask is None:
        mask = img > tol
    return img[np.ix_(mask.any(1), mask.any(0))]
```
